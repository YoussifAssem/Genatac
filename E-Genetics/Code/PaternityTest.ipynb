```python
import pandas as pd


data = pd.read_csv('drive/MyDrive/Paternity/data.csv')
notData = pd.read_csv('drive/MyDrive/Paternity/not.csv')
data.head()
```

```python
def check(child, parent):
  if(child[0] == parent[0] or child[0] == parent[1] or child[1] == parent[0] or child[1] == parent[1]):
    return True
  else:
    return False  
```

```python
f = []
m = []
c1 = []
c2 = []
c3= []
for i in data['father']:
  f.append(i)
for i in data['mother']:
  m.append(i)
for i  in data['child1']:
  c1.append(i)
for i in data['child2']:
  c2.append(i)
for i in data['child3']:
  c3.append(i)

print('Father: ', f)
print('Mother: ', m)
print('Child 1 : ', c1)
print('Child 2 : ', c2)
print('Child 3 : ', c3)
child = ['', '', ''] 
val = ''
for i in range(len(notData['child1'])):
  if((check(c1[i], f[i]) and check(c1[i], m[i])) and (check(c2[i], f[i]) and check(c2[i], m[i])) and (check(c3[i], f[i]) and check(c3[i], m[i]))):
    child[0] = 'Child1'
    child[1] = 'Child2'
    child[2] = 'Child3'
  else:
       child[0] = 'Not'
       child[1] = 'Not'
       child[2] = 'Not'
       break       


if(child[0] == 'Child1' and child[1] == 'Child2' and child[2] == 'Child3'):
      print('All Childs is related to this family')

elif(child[0] == 'Not'):
  print('Their are a child is not related to this family')

        
```
